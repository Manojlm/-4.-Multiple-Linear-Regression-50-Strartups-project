### NOTE:

```python
Companys.head(10)
```

## Model Building

```python
X=Companys.drop(['Profit'],axis=1)
Y=Companys[['Profit']]
```

```python
X.columns
```

```python
#Standardizing the data
from sklearn.preprocessing import StandardScaler
std_scaler=StandardScaler()
X_scaled=std_scaler.fit_transform(X)
X_scaled=pd.DataFrame(data=X_scaled,columns=X.columns)
X_scaled
```

```python
X_scaled.mean()
```

```python
X_scaled.std()
```

## Model Training

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X_scaled,Y)
```

```python
model.intercept_
```

```python
model.coef_
```

## Model Testing

```python
y_pred=model.predict(X_scaled)
```

## Model Evaluation

```python
error=Y - y_pred
error
```

```python
plt.scatter(x = Companys['Spend'],y = error)
plt.title('Homoscedasticity Check - R&D Spend Vs Error')
plt.xlabel('R&D Spend')
plt.ylabel('Error')
plt.show()
```

```python
plt.scatter(x = Companys['Administration'],y = error)
plt.title('Homoscedasticity Check - Administration Vs Error')
plt.xlabel('Administration')
plt.ylabel('Error')
plt.show()
```

```python
plt.scatter(x = Companys['Marketing'],y = error)
plt.title('Homoscedasticity Check - Marketing Spend Vs Error')
plt.xlabel('Marketing Spend')
plt.ylabel('Error')
plt.show()
```

#### Homoscedasity test is failed

## Zero residual Mean Test
It is a Plot between ***Actual output  v/s Predicted Output***

```python
plt.scatter(x = Y,y = y_pred)
plt.title('Actual Output Vs Predicted Output')
plt.xlabel('Actual Output')
plt.ylabel('Predicted Output')
plt.show()
```

## Model Building | Model Training using Statsmodels

```python
import statsmodels.formula.api as smf
```

```python
model=smf.ols(formula='Profit~Spend+Administration+Marketing',data=Companys).fit()
```

```python
model.params
```

```python
model.pvalues
```

```python
model_1 = smf.ols(formula = 'Profit~Spend', data = Companys).fit()
print('R-square                           : ',round(model_1.rsquared,4))
print('Adjusted R-square                  : ',round(model_1.rsquared_adj,4))
print('Akaike information criterion (AIC) : ',round(model_1.aic,4))
print('Bayesian information criterion(BIC): ',round(model_1.bic,4))
```

```python
model_2 = smf.ols(formula = 'Profit~Spend+Administration', data = Companys).fit()
print('R-square                           : ',round(model_2.rsquared,4))
print('Adjusted R-square                  : ',round(model_2.rsquared_adj,4))
print('Akaike information criterion (AIC) : ',round(model_2.aic,4))
print('Bayesian information criterion(BIC): ',round(model_2.bic,4))
```

```python
model_3 = smf.ols(formula = 'Profit~Spend+Administration+Marketing', data = Companys).fit()
print('R-square                           : ',round(model_3.rsquared,4))
print('Adjusted R-square                  : ',round(model_3.rsquared_adj,4))
print('Akaike information criterion (AIC) : ',round(model_3.aic,4))
print('Bayesian information criterion(BIC): ',round(model_3.bic,4))
```

```python
model_4 = smf.ols(formula = 'Profit~Spend+Marketing', data = Companys).fit()
print('R-square                           : ',round(model_4.rsquared,4))
print('Adjusted R-square                  : ',round(model_4.rsquared_adj,4))
print('Akaike information criterion (AIC) : ',round(model_4.aic,4))
print('Bayesian information criterion(BIC): ',round(model_4.bic,4))
```

#### Model 4 is the best model

## Model Building using sklearn library

### 2 Steps approch
    1.Separate your X and Y
    2.Model validation technique
        *Train-test split
        *k-Fold Cv
        *Loocv

```python
X=Companys.drop(['Profit'],axis=1)
Y=Companys[['Profit']]
```

```python
X.head()
```

```python
Y.head()
```

```python
X.shape,Y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.20,shuffle=True,random_state=None)
```

```python
x_train.shape,y_train.shape
```

```python
x_train
```

```python
x_test.shape,y_test.shape
```

```python
y_train
```

## Model Training

```python
from sklearn.linear_model import LinearRegression
linear_model=LinearRegression()
linear_model.fit(x_train,y_train)
```

## Model Testing | Model Evaluation

##### Calucate the error for Training data

```python
y_pred_train=linear_model.predict(x_train)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
mean_squared_error(y_train,y_pred_train)
```

```python
mean_absolute_error(y_train,y_pred_train)
```

#### Calculate the Error for the Test data

```python
y_pred_test=linear_model.predict(x_test)
```

```python
mean_squared_error(y_test,y_pred_test)
```

```python
mean_absolute_error(y_test,y_pred_test)
```
